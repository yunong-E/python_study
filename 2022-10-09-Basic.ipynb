```python
# 데이터의 특징을 알 수 있다. 
# 컬럼명, row와 columns의 갯수, 데이터 타입, 결측치 갯수 ... 등

# RangeIndex: row의 갯수
# Data columns: columns의 갯수

# 결측치 확인 법: Non-Null Count 확인.
# 12 non-null 의 의미: 결측값이 없다.
# 예를 들어, 결측값이 3개 존재한다면 9 non-null 로 확인이 되었을 것.

sl1.info()
```

```python
sl1.describe()
```

```python
# 중복 데이터의 갯수 알아보기.
sl1.duplicated().sum()
```

```python
# 중복데이터를 만들기 위해서 임의로 데이터를 추가.
# append 이용시
new_data = {
    '구분' : '2014년8월',
    '사고(건)' : 222,
  	'사망(명)' : 9,
    '부상(명)' : 447
}

sl1 = sl1.append(new_data, ignore_index = True)
sl1
```

```python
# loc 이용
new_data2 = ['2014년4월', 182, 4, 360]

sl1.loc[len(sl1)] = new_data2
sl1
```

```python
# 중복값 재확인
sl1.duplicated().sum()
```

```python
# 중복값 제거 후
sl1 = sl1.drop_duplicates()

# 중복값이 제거된 것을 확인
sl1.duplicated().sum()

# 중복값 제거 후, index가 뒤죽박죽이 되었다면 재정렬해주기
# drop = True 옵션 적용 시, 기존 인덱스를 버리고 재배열 한다.
sl1 = sl1.reset_index(drop = True)
```

```python
sl1
```

```python
# Feature Engineering
# 기존의 데이터를 이용하여 의미있는 새로운 데이터를 만들어내는 것

# 새로운 데이터프레임 추가
from google.colab import files
uploaded = files.upload()
```

```python
sl2 = pd.read_excel('2015년 졸음운전 교통사고.xlsx')
sl2
```

```python
# 두 데이터프레임을 합치기 전, 몇 년도 사고인지 구분이 가능한 array 생성하기
sl_2014 = np.repeat(2014, len(sl1))
sl_2014
```

```python
# np.repeat(인수1, 인수2)
sl_2015 = np.repeat(2015, len(sl2))
sl_2015
```

```python
# 위에서 만든 array를 기존 데이터 프레임에 추가하기
sl1['년도'] = sl_2014
sl2['년도'] = sl_2015

# 새로운 컬럼 및 데이터가 추가 되었는지 확인하기.
sl1.columns
```

```python
# 두 데이터프레임의 columns 수가 동일한지 확인
set(sl1.columns) == set(sl2.columns)
```

```python
sl = sl1.append(sl2, ignore_index = True)
print(sl.shape)
```

```python
sl
```

```python
# index = False 로 지정해야 첫 번째 컬럼이 인덱스로 지정이 된다.
# 그런데 어디에 저장이 되나요?
sl.to_csv('2014년_2015년_졸음운전.csv', index = False)
```
